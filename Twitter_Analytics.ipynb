```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import collections
from collections import Counter

# Reads the twitter data
df = pd.read_json('tweet_data.json')
# Removes unneccessary data
df_ic = df.drop(columns=['extended_entities','retweeted','favorited','display_text_range','truncated','id_str','id','source','in_reply_to_user_id_str','in_reply_to_user_id','in_reply_to_status_id_str','in_reply_to_status_id','user','geo','coordinates','place','contributors','lang'])
df_ic
```

```python
# Average Likes
print('Average Likes:',df_ic['favorite_count'].mean())

# Average Retweets
print('Average Retweets:',df_ic['retweet_count'].mean())

# Max Likes
print('Maximum Likes:',df_ic['favorite_count'].max())

# Max Retweets
print('Maximum Retweets:',df_ic['retweet_count'].max())
```

```python
import datetime
from datetime import datetime

date_1 = df['created_at'].iloc[0]
date_2 = df['created_at'].iloc[-1]

days = date_1 - date_2
tweets_per_day = len(df) / days.days
tweets_per_hour = tweets_per_day / 24
print("Tweets per day:", tweets_per_day)
print("Tweets per hour:",tweets_per_hour)
```

```python
# Best Tweet
idx = df_ic['favorite_count'].argmax()
print('Most liked:\n',df_ic['full_text'].iloc[idx])
```

```python
# Most Retweeted
idx = df_ic['retweet_count'].argmax()
print('Most Retweeted:\n', df_ic['full_text'].iloc[idx])
```

```python
# Hashtag Counter
hashtags = []
for i in range(len(df_ic)):
  if df_ic['entities'][i]['hashtags'] != []:
    for hashtag in df_ic['entities'][i]['hashtags']:
      hashtags.append(hashtag["text"])
hashtag_count = Counter()
for i in hashtags:
  hashtag_count[i] += 1

sorted_keys = sorted(hashtag_count.keys(), key=lambda key: hashtag_count[key], reverse=True)

for key in sorted_keys:
  print(key, hashtag_count[key])
```

```python
# Highest Favorited Tweets and their hashtags
liked_sorted = df_ic.sort_values(by = 'favorite_count', ascending = False)
top_posts = liked_sorted.copy()
hash_list = []
for index, rows in liked_sorted.iterrows():
  hashtags = []
  for i in rows['entities']['hashtags']:
    hashtags.append(i['text'])
  hash_list.append(hashtags)

top_posts = top_posts.drop(columns=['created_at','entities','possibly_sensitive','retweeted_status','quoted_status_id','quoted_status_id_str','quoted_status_permalink','quoted_status'])
top_posts.insert(1,"HashTags",hash_list, True)
top_posts
```

```python
# Highest Favorited Tweets and their hashtags
retweet_sorted = df_ic.sort_values(by = 'retweet_count', ascending = False)
top_retweet = retweet_sorted.copy()
hash_list = []
for index, rows in retweet_sorted.iterrows():
  hashtags = []
  for i in rows['entities']['hashtags']:
    hashtags.append(i['text'])
  hash_list.append(hashtags)

top_retweet = top_retweet.drop(columns=['created_at','entities','possibly_sensitive','retweeted_status','quoted_status_id','quoted_status_id_str','quoted_status_permalink','quoted_status'])
top_retweet.insert(1,"HashTags",hash_list, True)
top_retweet
```

```python
# Quoted Tweets based on Likes
quoted_liked = liked_sorted[liked_sorted["is_quote_status"]==True].drop(columns=['created_at','possibly_sensitive','retweeted_status','quoted_status_id','quoted_status_id_str','quoted_status_permalink','quoted_status','in_reply_to_screen_name'])
quote_like = quoted_liked.copy()
hash_list = []
for index, rows in quoted_liked.iterrows():
  hashtags = []
  for i in rows['entities']['hashtags']:
    hashtags.append(i['text'])
  hash_list.append(hashtags)

quote_like = quote_like.drop(columns=['entities','is_quote_status'])
quote_like.insert(1,"HashTags",hash_list, True)
quote_like
```

```python
# Quoted Tweets based on Retweets
quoted_retweet = retweet_sorted[retweet_sorted["is_quote_status"]==True].drop(columns=['created_at','possibly_sensitive','retweeted_status','quoted_status_id','quoted_status_id_str','quoted_status_permalink','quoted_status','in_reply_to_screen_name'])
quote_retweet = quoted_retweet.copy()
hash_list = []
for index, rows in quoted_retweet.iterrows():
  hashtags = []
  for i in rows['entities']['hashtags']:
    hashtags.append(i['text'])
  hash_list.append(hashtags)

quote_retweet = quote_retweet.drop(columns=['entities','is_quote_status'])
quote_retweet.insert(1,"HashTags",hash_list, True)
quote_retweet
```

```python
# Replies that got good likes
top_posts.dropna().drop(columns=['is_quote_status'])
```

```python
# Replies that got good retweets
top_retweet.dropna().drop(columns=['is_quote_status'])
```

```python
form = []

s = top_posts['full_text']
for i in s:
  interrogative = i.count('?')
  exclamitory = i.count('!')
  declaritve = i.count('.')
  punctuation = np.array([interrogative, exclamitory, declaritve])
  if punctution[0] > 1:
    form.append(0)
  index_max = np.argmax(punctuation)
  form.append(index_max)

top_posts.insert(2,"Sentence",form, True)
top_posts['Sentence'] = np.where((top_posts.Sentence == '0'),'Interrogative',top_posts.Sentence)
top_posts['Sentence'] = np.where((top_posts.Sentence == '1'),'Exclamitory',top_posts.Sentence)
top_posts['Sentence'] = np.where((top_posts.Sentence == '2'),'Declarative',top_posts.Sentence)
top_posts
```
